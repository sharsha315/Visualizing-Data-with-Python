## 3. Visualizing the data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Select numerical columns of type float
float_columns = titanic_df.select_dtypes(include=['float64'])
```

```python
float_columns.head()
```

```python
# replacing rows with missing values with mean values

for col in float_columns.columns:
  # calculate the mean of each column
  col_mean = titanic_df[col].mean()

  # Fill missing values in each column with its mean
  float_columns[col].fillna(col_mean, inplace=True)
```

### Visualizing numerical columns Age and Fare
### 1. Histograms

```python
for col in float_columns.columns:
  plt.figure(figsize=(8, 6))
  sns.histplot(float_columns[col], kde=True)
  plt.xlabel(col)
  plt.ylabel('Frequency')
  plt.title(f'Distribution of {col}')
  plt.show()
```

### 2. Density plots

```python
for col in float_columns.columns:
  plt.figure(figsize=(8, 6))
  sns.kdeplot(float_columns[col], fill=True)
  plt.xlabel(col)
  plt.ylabel('Density')
  plt.title(f'Density plot of {col}')
  plt.show()
```

### 3. Box Plots

```python
for col in float_columns.columns:
  sns.boxplot(float_columns[col])
  plt.xlabel(col)
  plt.title(f'Box plot of {col}')
  plt.show()
```

### 4. Pair plot for Age and Fare

```python
sns.pairplot(float_columns)
plt.show()
```

### 5. Scatter Plot

```python
plt.scatter(float_columns['Age'],float_columns['Fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot')
plt.show()
```

### Visualizing Categorical columns Sex and Embarked
### 1. Bar Plot

```python
cat_columns = titanic_df[['Sex','Embarked']]
```

```python
for col in cat_columns.columns:
  cat_columns[col].value_counts().plot(kind='bar')
  plt.xlabel(col)
  plt.ylabel('Count')
  plt.title(f'Bar plot of {col}')
  plt.show()
```

### 2. Pie Charts

```python
for col in cat_columns.columns:
  cat_columns[col].value_counts().plot(kind='pie', autopct='%1.1f%%')
  plt.ylabel('')
  plt.title(f'Pie chart of {col}')
  plt.show()
```
